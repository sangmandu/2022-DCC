# 전처리

전처리시 해야되는 작업은 다음과 같다.

1. 나이의 결측값을 채워줘야 한다.
2. 성별을 0과 1로 매핑시켜야 한다.
3. PassengerId,Ticket,Cabin,Name열을 지워줘야 한다. 학습하기에 부적절한 컬럼들이기 때문이다.


```python
train['Age'] = train['Age'].fillna(29)
train['Sex'] = train['Sex'].map({'male' : 0, 'female' : 1})

test['Age'] = test['Age'].fillna(29)
test['Sex'] = test['Sex'].map({'male' : 0, 'female' : 1})
```

```python
train_x = train[['Pclass', 'SibSp','Age', 'Sex',"Fare","Parch"]]
train_y = train['Survived']

test_x = test[['Pclass', 'SibSp','Age', 'Sex',"Fare","Parch"]]

#결측 행 지우기
test_x["Fare"] = test_x["Fare"].fillna(0)
```

# 모델링

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## # 모델 만들기
model = DecisionTreeClassifier()
# 학습하기
model.fit(train_x, train_y)
test_y_pred = model.predict(test_x)
```

## 제출 파일 생성

```python
submission = pd.read_csv("data/submission.csv")
submission['Survived'] = test_y_pred
submission.to_csv('submit/DecisionTreeClassifier_종원.csv', index = False)
```

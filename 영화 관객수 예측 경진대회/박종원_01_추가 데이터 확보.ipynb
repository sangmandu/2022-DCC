```python
#일정기간 내 영화 정보 수집
import requests
import json
import time
import datetime
import pandas as pd

i = 0
my_datetime = datetime.datetime(2010, 1, 1)
l = []
while True:
    new_datetime = my_datetime + datetime.timedelta(days=i)
    targetDt = new_datetime.strftime('%Y%m%d') #20180220
    #print(targetDt)
    #url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=130ad25d8935a673ca47beb43436aa85&targetDt=20190401'
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=130ad25d8935a673ca47beb43436aa85&targetDt='+targetDt
    res = requests.get(url)
    text = res.text
    #print(text)

    d = json.loads(text) 
    #print(d)

    print(d['boxOfficeResult']['showRange'])

    for b in d['boxOfficeResult']['dailyBoxOfficeList']:
        print(new_datetime.strftime('%Y-%m-%d'),b['rankOldAndNew'],b['salesAmt'], b['movieCd'], b['movieNm'], b['showCnt'], b['scrnCnt'], b['openDt'], b['audiAcc'])
        l.append([new_datetime.strftime('%Y-%m-%d'),b['rankOldAndNew'],b['salesAmt'], b['movieCd'], b['movieNm'], b['showCnt'], b['scrnCnt'], b['openDt'], b['audiAcc']])
    time.sleep(1)
    i = i + 1

    if targetDt >= '20160101':
        break
df = pd.DataFrame(l, columns=['date', 'rankOldAndNew', 'salesAmt' , 'movieCd', 'movieNm', 'showCnt', 'scrnCnt', 'openDt', 'audiAcc'])

df.to_csv('kobis_movies2.csv', index=False ,encoding='CP949') 
```
